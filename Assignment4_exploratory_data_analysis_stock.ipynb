## Summarize the information stored in the data
All the stock tickers details are stored in the companylist.csv file,
All the stock related data are stored in data frame using yfinance library.Historical Stock Data has all the details about instituional holders, stock's open price, close price,high and low along with volume data.

## Make a note of any documentation about the data
There is no null values in the data.Hisotrical data consists of 7 columns and 10491 rows.


```python
#importing pandas,matplot lib and yfinance python libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


#defining pandas data frame to read ticker symbols from the companylist.csv
df=pd.read_csv('companylist.csv')
#print(df['Symbol'])

#fethcing historical stock data using yfinance library
aapl = yf.Ticker('AAPL')
#aapl.info
hist = aapl.history(period="max")


#printing first few rows of the data
print(hist.head())

#printing last few rows of the data
print(hist.tail())



#describing the data 
print(hist.describe())

# show major holders
print (aapl.major_holders.head())

# show institutional holders
print(aapl.institutional_holders)


#checking for Null values in the data
print(hist.isnull())

#ploting volume vs Closing price data and 
volume = hist['Volume']

close = hist['Close']
plt.plot(close,volume)
plt.show()

```
